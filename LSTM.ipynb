```python
def sigmoid_activation(x):
    return 1 / (1 + (2.718281828459045 ** -x))  

def tanh_activation(x):
    return (2 / (1 + (2.718281828459045 ** (-2 * x)))) - 1  


input_sequence = [1, 2, 3]

# forget, input, candidate, and output gates
Wf, Whf, bf = 0.5, 0.1, 0
Wi, Whi, bi = 0.6, 0.2, 0
Wc, Whc, bc = 0.7, 0.3, 0
Wo, Who, bo = 0.8, 0.4, 0
Wy, by = 4, 0

# hidden state (h) and cell state (C)
hidden_state = 0.0
cell_state = 0.0


print("LSTM Forward Pass Simulation")
print(f"Input sequence: {input_sequence}")
print(f"Initial h0 = {hidden_state}, C0 = {cell_state}\n")

for time_step, input_value in enumerate(input_sequence, start=1):
    # Calculate forget gate (f), input gate (i), candidate cell state (C_tilde), and output gate (o)
    forget_gate = sigmoid_activation(Wf * input_value + Whf * hidden_state + bf)
    input_gate = sigmoid_activation(Wi * input_value + Whi * hidden_state + bi)
    candidate_cell = tanh_activation(Wc * input_value + Whc * hidden_state + bc)

    # Update 
    cell_state = forget_gate * cell_state + input_gate * candidate_cell

    # Calculate output gate and hidden state
    output_gate = sigmoid_activation(Wo * input_value + Who * hidden_state + bo)
    hidden_state = output_gate * tanh_activation(cell_state)

    # Output results for this time step
    print(f"Time Step {time_step}")
    print(f"Forget gate (f) = {round(forget_gate, 3)}")
    print(f"Input gate (i) = {round(input_gate, 3)}")
    print(f"Candidate cell (C_tilde) = {round(candidate_cell, 3)}")
    print(f"Cell state (C) = {round(cell_state, 3)}")
    print(f"Output gate (o) = {round(output_gate, 3)}")
    print(f"Hidden state (h) = {round(hidden_state, 3)}\n")

# Predicting at time step 4 (input value = 4)
next_input = 4
f4 = sigmoid_activation(Wf * next_input + Whf * hidden_state + bf)
i4 = sigmoid_activation(Wi * next_input + Whi * hidden_state + bi)
C_tilde4 = tanh_activation(Wc * next_input + Whc * hidden_state + bc)
C4 = f4 * cell_state + i4 * C_tilde4
o4 = sigmoid_activation(Wo * next_input + Who * hidden_state + bo)
h4 = o4 * tanh_activation(C4)

# Show prediction results for time step 4
print("Prediction at Time Step 4")
print(f"f4 = {round(f4, 3)}")
print(f"i4 = {round(i4, 3)}")
print(f"C_tilde4 = {round(C_tilde4, 3)}")
print(f"C4 = {round(C4, 3)}")
print(f"o4 = {round(o4, 3)}")
print(f"h4 = {round(h4, 3)}\n")

# Final output prediction (y_hat)
final_prediction = Wy * h4 + by
print("Final Prediction")
print(f"Predicted Output (y_hat) = {round(final_prediction, 3)}")
```
